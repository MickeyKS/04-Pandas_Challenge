```python
#Dependencies
import pandas as pd 

#Load CSV File
heroesofpymoli = "Resources/purchase_data.csv"

#Read Purchasing Data File and store into Pandas DataFrame
purchase_data = pd.read_csv(heroesofpymoli)

#Correct SN to Screen Name in Columns
purchase_data.rename(columns={"SN":"Screen Name"}, inplace=True)

#Print DataFrame
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
#Total Number of Players
player = len(purchase_data["Screen Name"].unique())

#Create DataFrame
player_count = pd.DataFrame({"Total Players":[player]})

#Print Player Count
player_count.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of Unique Items
unique_items = len(purchase_data["Item Name"].unique())

#Average Price of Purchases
average_price = purchase_data["Price"].mean()

#Total Number of Purchases
total_purchases = len(purchase_data["Item Name"])

#Total Revenue
total_revenue = purchase_data["Price"].sum()

#Create DataFrame
purchasing_analysis = pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":[average_price], "Total Purchases":[total_purchases], "Total Revenue":[total_revenue]})

#Format DataFrame
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].astype(float).map("${:,.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)

#Print Purchasing Analysis
purchasing_analysis.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Set Data to only Count First Iteration of Screen Name
count_first = purchase_data.drop_duplicates("Screen Name")
total_genders = count_first["Gender"].count()

#Total Count by Gender
gender_male = count_first["Gender"].value_counts()["Male"]
gender_female = count_first["Gender"].value_counts()["Female"]
gender_other = total_genders - gender_male- gender_female

#Percentage of Players by Gender
male_percentage = (gender_male/total_genders)*100
female_percentage = (gender_female/total_genders)*100
other_percentage = (gender_other/total_genders)*100

#Create DataFrame
gender_count = pd.DataFrame({"": ["Male", "Female", "Other/Non-Disclosed"], "Percentage of Players": [male_percentage, female_percentage, other_percentage], "Total Count": [gender_male, gender_female, gender_other]})

#Format Data
gender_count["Percentage of Players"] = gender_count["Percentage of Players"].map("{:.2f}%".format)

#Print Gender Demographics
gender_count.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by Gender and Screen Name
gender_grouped = purchase_data.groupby(["Gender"])
count_first = purchase_data.drop_duplicates("Screen Name")
total_genders = count_first["Gender"].count()

#Purchase Count
purchase_count = gender_grouped["Screen Name"].count()

#Average Purchase Price
average_price = gender_grouped["Price"].mean()

#Total Purchase Value
purchase_value = gender_grouped["Price"].sum()

#Average Purchase Totals by Gender
purchases_by_gender = (gender_grouped["Price"].sum()/gender_grouped["Screen Name"].count())

#Create DataFrame
purchasing_totals = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_price, "Total Purchase Value": purchase_value, "Average Purchase Totals by Gender": purchases_by_gender})

#Format DataFrame
purchasing_totals["Average Purchase Price"] = purchasing_totals["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchasing_totals["Total Purchase Value"] = purchasing_totals["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchasing_totals["Average Purchase Totals by Gender"] = purchasing_totals["Average Purchase Totals by Gender"].astype(float).map("${:,.2f}".format)

#Print DataFrame
purchasing_totals.head()
```

## Age Demographics

* Establish bins for ages

* Categorize the existing players using the age bins. Hint: use pd.cut()

* Calculate the numbers and percentages by age group

* Create a summary data frame to hold the results

* Optional: round the percentage column to two decimal points

* Display Age Demographics Table

```python
# Categorize the existing players using the age bins 
bins = [0,9,14,19,24,29,34,39,40]
age_bins = ["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]
```

```python
#Place Data
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins=bins, labels=age_bins)

#Group by Age Range
age_range = purchase_data.groupby("Age Range")

#Age Totals
age_total = age_range["Screen Name"].nunique()

#Age Percentages
age_percentage = round((age_total/576) * 100,2)

#Create DataFrame
age_demographics = pd.DataFrame({"Total Count": age_total, "Percentage of Players": age_percentage})

#Format Data
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{0:,.2f}%".format)

#Print DataFrame
age_demographics.head(8)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by Screen Name
top_spenders = purchase_data.groupby("Screen Name")

#Purchase Count
spender_purchase_count = top_spenders["Screen Name"].count()

#Average Purchase Price
spender_average_price = round(top_spenders["Price"].mean(),2)

#Total Purchase Value
spender_total_value = top_spenders["Price"].sum()

#Create DataFrame
top_five_spenders = pd.DataFrame({"Purchase Count": spender_purchase_count, "Average Purchase Price": spender_average_price, "Total Purchase Value": spender_total_value})

#Sort Total Purchases in Descending Order
top_five_spenders = top_five_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

#Format DataFrame
top_five_spenders["Average Purchase Price"] = top_five_spenders["Average Purchase Price"].astype(float).map("${:,.2f}".format)
top_five_spenders["Total Purchase Value"] = top_five_spenders["Total Purchase Value"].astype(float).map("${:,.2f}".format)

#Print Top Five Spenders
top_five_spenders.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#List Item ID, Item Name, and Item Price
popular_items_list = purchase_data[["Item ID", "Item Name", "Price"]]

#Group by Item ID & Item Name
popular_items = popular_items_list.groupby(["Item ID", "Item Name"])

#Purchase Count
item_purchase_count = popular_items["Price"].count()

#Item Price
item_price = popular_items["Price"].sum()

#Total Purchase Value
item_purchase_value = item_price / item_purchase_count

#Create DataFrame
most_popular_items = pd.DataFrame({"Purchase Count": item_purchase_count, "Item Price": item_purchase_value, "Total Purchase Value": item_price})

#Sort Total Purchases in Descending Order
most_popular_items = most_popular_items.sort_values(["Total Purchase Value"], ascending=False).head()

#Format DataFrame
most_popular_items["Total Purchase Value"] = most_popular_items["Total Purchase Value"].astype(float).map("${:,.2f}".format)
most_popular_items["Item Price"] = most_popular_items["Item Price"].astype(float).map("${:,.2f}".format)

#Print Most Popular Items
most_popular_items.head(5)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort Most Profitable Items in Descending Order
most_profitable = most_popular_items.sort_values(["Total Purchase Value"], ascending=False).head()

#Print DataFrame
most_profitable.head(5)
```
